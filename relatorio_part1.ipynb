# Relatório Inicial – Projeto de Introdução à Ciência de Dados

## Tema

O tema escolhido para este projeto é **Acidentes de Trânsito no Brasil**, utilizando como base o conjunto de dados disponibilizado pelo **RENAEST** (Registro Nacional de Acidentes e Estatísticas de Trânsito).

O objetivo é **compreender padrões e fatores associados à gravidade dos acidentes**, considerando variáveis relacionadas a veículos, vítimas, localidades e circunstâncias do acidente.

---

## Conjunto de Dados

O **RENAEST** é dividido em quatro arquivos principais, interligados por identificadores de acidentes:

| Arquivo | Conteúdo Principal |
| :--- | :--- |
| `Acidentes_DadosAbertos` | Informações gerais do acidente: data, hora, tipo de via, condições climáticas e número de vítimas. |
| `Vitimas_DadosAbertos` | Dados sobre as vítimas envolvidas: idade, sexo e **gravidade** (ileso, ferido leve, ferido grave, óbito). |
| `TipoVeiculo_DadosAbertos` | Tipo de veículo envolvido (automóvel, motocicleta, caminhão, bicicleta etc.). |
| `Localidade_DadosAbertos` | Características do local do acidente: município, estado e rodovia. |

Este conjunto de dados é abrangente e permite analisar tanto os fatores ambientais quanto os humanos e veiculares.

---

## Perguntas de Pesquisa

As análises terão como foco responder às seguintes questões **simplificadas**:

1.  **Motocicletas têm mais acidentes com óbitos que automóveis?**
    * *Comparação direta:* Proporção de óbitos entre motocicletas vs automóveis
    * *Atributos:* `tipo_veiculo`, `qtde_obitos`
2.  **Finais de semana têm mais acidentes graves que dias úteis?**
    * *Comparação temporal:* Acidentes com óbito em fim de semana vs dias úteis  
    * *Atributos:* `dia_semana`, `qtde_obitos`
3.  **Jovens (18-34 anos) se acidentam mais?**
    * *Foco etário:* Jovens vs outras idades em condições ruins de pista
    * *Atributos:* `faixa_idade`, `cond_pista`, `qtde_obitos`
4.  **Rodovias federais têm mais óbitos?**
    * *Análise ambiental:* Tipo de via vs condições climáticas
    * *Atributos:* `tp_rodovia`, `cond_meteorologica`, `qtde_obitos`

---

## Planejamento dos Próximos Passos

1.  **Importação e limpeza dos dados:** Carregamento dos dados + tratamento de valores "não informado/desconhecido" + criação de variáveis derivadas.
2.  **Análise exploratória (EDA):** Visualizações detalhadas das estruturas de dados para cada pergunta + estatísticas descritivas.
3.  **Testes de hipótese:** Aplicação de testes estatísticos simples (Z-test, Qui-quadrado) para responder cada pergunta.
4.  **Relatório final:** Síntese dos achados com visualizações e conclusões diretas.
